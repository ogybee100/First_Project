```python
#No 2c
# Calculating the mean AI usage for each sector
mean_ai_usage = df[[f'AI Use in Sector {i} (%)' for i in range(1, 49)]].mean()

# Sorting the sectors by mean AI usage in descending order
top_sectors = mean_ai_usage.sort_values(ascending=False).head(5)

# Print the top 5 sectors
print("Top 5 Sectors with Highest AI Usage:")
print(top_sectors)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#No 4b
# Selecting columns corresponding to AI use in each sector
ai_use_columns = [f'AI Use in Sector {i} (%)' for i in range(1, 49)]
ai_use_df = df[ai_use_columns]

# Calculating the correlation matrix
correlation_matrix = ai_use_df.corr()

# Creating a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Heatmap for AI Use Between Different Sectors')
plt.show()
```
